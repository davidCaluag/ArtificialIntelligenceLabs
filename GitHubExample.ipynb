```
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
import pandas as pd

pd = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)
pd.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
pd.head()

pd.drop('class', axis=1, inplace=True)
pd.head()
#Just petal
pd.drop('sepal_length', axis=1, inplace=True)
pd.head()
pd.drop('sepal_width', axis=1, inplace=True)
pd.head()
```

```
#your codes here

X = pd[['petal_width', 'petal_length']].values

#plt.scatter(pd['petal_width'], pd['petal_length'], s=50);
plt.scatter(X[:, 0], X[:, 1], s=50);


from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=8)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=1, random_state = 0)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

```
Sum_of_squared_distances = []
K = range(1,10)
for k in K:
 km = KMeans(n_clusters=k)
 km = km.fit(X)
 Sum_of_squared_distances.append(km.inertia_)

for k in range(1,9):
  print(k, Sum_of_squared_distances[k])

plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

The Elbow Method suggests 3 as the optimal number of clusters. The method involves examining the Sum of Squared Distances (SSD), also called Within-Cluster Sum of Squares (WCSS), which measures the squared distance between each point and its assigned cluster center.

With k=1, the SSD is high because all points are forced into one cluster.
At k=2, there’s a significant drop in SSD, but k=3 reduces it further to a stable point.
From k=4 onward, SSD decreases only minimally, creating a subtle “elbow” at k=3 where the SSD stops dropping sharply. This indicates that k=3 provides a balance between compact clusters and model simplicity.
Thus, k=3 is optimal, as adding more clusters doesn’t significantly reduce the SSD, showing diminishing returns beyond this point.
